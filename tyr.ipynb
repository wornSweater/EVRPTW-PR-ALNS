```python
p = 0.01
n = 0.5
N = 1000

folder_path = "evrptw_instances"

files_1 = glob(f'{folder_path}/*C5.txt')
files_2 = glob(f'{folder_path}/*C10.txt')
files_3 = glob(f'{folder_path}/*C15.txt') 
files_4 = glob(f'{folder_path}/*21.txt') 

files = files_1 + files_2 + files_3 + files_4

for p in tqdm([0.1, 0.05, 0.01]):
    
    for n in [1, 0.5, 0.1]:
        
        out_out = [["name", "feasibility", "p", "n"]]
        
        for file in tqdm(files):
            
            count = 0
            
            parameters = get_parameters(file)
            mip = MIPCheck(parameters)
            helper = Helper(parameters)
            with open(Path(file).stem + ".json", 'r') as input:
                data = json.load(input)
            
            mip.update_times(p, n)
            
            for _ in range(N):
                mip.update_times(p, n)
                if helper.feasible(data):
                    count += 1
                    
            out_out.append([Path(file).stem, count/N, p, n])
    
        with open(f'Feasibility_{p}_{n}.csv', 'w', newline='') as op:
            writer = csv.writer(op)
            writer.writerows(out_out)
```
